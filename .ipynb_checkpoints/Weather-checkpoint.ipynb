```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import csv
from datetime import datetime
```

```python
#read the file
fare_data = pd.read_csv("sample.csv")
temp_data=pd.read_csv("temperature.csv")
temp_data = temp_data[['date','New York']]
temp_data.head()
```

```python
#check data type
temp_data.dtypes
```

```python
#convert date from strings to timestamp
temp_data['timestamp'] = [datetime.strptime(d, "%Y-%m-%d %H:%M:%S") for d in temp_data["date"]]
temp_data.head()
fare_data.head()
```

```python
#clean fare_data time format
fare_data['new_time'] = (fare_data['pickup_datetime']).apply(lambda x: datetime.strptime(x, "%Y-%m-%d %H:%M:%S %Z"))    
fare_data = fare_data.drop(['Unnamed: 0','key','pickup_datetime'], axis=1)
time_range = (fare_data['new_time'] > "2013-01-01") & (fare_data['new_time'] < "2016-12-31")
ranged_data = fare_data.loc[time_range]
ranged_data.tail()
```

```python
#grab the hour of the fare_data
temp_data['new_date']=temp_data.timestamp.dt.date
temp_data['new_hour']=temp_data.timestamp.dt.hour

temp_data.head()
```

```python

combined_fare_temp_df = pd.merge(fare_data, temp_data[['date','New York']],
                                 how='left', on='new_date' and 'new_hour')
#combined_fare_temp_df.head()
```
